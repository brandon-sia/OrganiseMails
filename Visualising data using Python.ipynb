# Data visualised for better decision making

```python
import os 
import pandas as pd
import tkinter as tk
from tkinter import filedialog 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Opens a window to choose file for data analysis

```python
root = tk.Tk()


root.lift()
root.attributes("-topmost", True)
prompt = tk.Label(text="Choose file to analyse data")
prompt.config(font=("Arial", 50))
prompt.pack()


csv_file = filedialog.askopenfilename()

root.withdraw()
root.destroy()
```

```python
df = pd.read_excel(csv_file)
```

```python
df.head()
```

## Check out which columns are in the datasets 

```python
df.columns
```

## Determine data type of columns, number of rows and columns 

```python
df.info()
```

## Visualise relationships with each variable 

```python
plt.figure(figsize=(10,10))
sns.pairplot(df)
```

## Correlation of the variables

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cmap="viridis",annot=True)
```

## Total sales and profit of brands in each country from year 2013-2014

```python
pd.pivot_table(df,index=["Product","Year","Country"],values=[' Sales','Profit'],aggfunc=np.sum)
```

```python
sale_map = pd.pivot_table(df,index=["Product","Year","Country"],values=[' Sales','Profit'],aggfunc=np.sum)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(sale_map,cmap="viridis",annot=True)
```

## Average sales and profit of brands in each country from year 2013-2014

```python
pd.pivot_table(df,index=["Product","Year","Country"],values=[' Sales','Profit'],aggfunc=np.mean)
```

```python
avg_map = pd.pivot_table(df,index=["Product","Year","Country"],values=[' Sales','Profit'],aggfunc=np.mean)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(avg_map,cmap="viridis",annot=True)
```

## Total sales visualised for brands in each country
#### It is easier to notice the best performing brands in terms of sales using a heatmap

```python
heatmap_data = pd.pivot_table(df,index=["Product","Country"],values=[' Sales'],aggfunc=np.sum)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(heatmap_data,cmap="coolwarm",annot=True)
```

# Now we move on to profit ratio analysis...

# Profit Margin of each brand

```python
df['Gross Profit Margin'] = df['Profit']/df[' Sales']
heatmap_data3 = pd.pivot_table(df,index=["Product"],values=['Gross Profit Margin'])
```

## Each brand has roughly the same profit margin ratio

```python
plt.figure(figsize=(10,10))
sns.heatmap(heatmap_data3,cmap="coolwarm",annot=True,vmin=0,vmax=0.5)
```

# Profit Margin of brands in each country

```python
heatmap_data4 = pd.pivot_table(df,index=["Product","Country"],values=['Gross Profit Margin'])
```

## From heatmap below, we can observe that Amarilla-US and Carretera-Canada have higher profit margins.

```python
plt.figure(figsize=(10,10))
sns.heatmap(heatmap_data4,cmap="coolwarm",annot=True,vmin=0,vmax=0.5)
```
